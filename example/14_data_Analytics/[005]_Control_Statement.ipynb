# I. 조건문(if ~ else)

> ## 1) if

* <font color = 'red'>조건문 뒤에 콜론(:) 붙임</font>
* <font color = 'red'>들여쓰기(Indentation) 주의</font>

```
score = 88
```

```
if score == 88:
    print('획득한 점수는 %d점 입니다.' % score)
```

> ## 2) if ~ else

```
if score == 100:
    print('100점 만점입니다.')
else:
    print('아쉽지만 만점은 아니네요.')
```

> ## 3) if ~ elif ~ else

* 사용자 입력

```
score = int(input('점수를 입력하세요: '))
```

* if 중첩

```
if score > 91:
    print('A')
else:
    if score > 81:
        print('B')
```

* elif 적용

```
if score > 91:
    print('A')
elif score > 81:
    print('B')
elif score > 71:
    print('C')
elif score > 61:
    print('D')
else:
    print('F')
```

# II. 반복문(for ~ in)

> ## 1) for ~ in with range( )

* range(Start, Stop, Step)

```
for i in range(1, 10, 2):
    print(i)
```

> ## 2) for ~ in with List_1

```
L1 = ['A', 'B', 'C', 'D', 'E']

for i in L1:
    print(i)
```

> ## 3) for ~ in with List_2

```
L2 = [['A', 11], ['B', 22], ['C', 33]]

for i, j in L2:
    print(i, 'and', j)
```

> ## 4) for ~ in with if ~ else

```
score = [99, 88, 55, 77, 66, 55, 100]
num = 0
```

```
for s in score: 
    num = num + 1

    if s > 60: 
        print('%d번 수강생은 수료입니다.' % num)
    else: 
        print('%d번 수강생은 미수료입니다.' % num)
```

* enumerate(리스트)
 - 순서가 있는 자료형의 값을 인덱스와 함께 전달(열거)

```
for num, s in enumerate(score):
    num = num + 1

    if s > 60: 
        print('%d번 수강생은 수료입니다.' % num)
    else: 
        print('%d번 수강생은 미수료입니다.' % num)
```

> ## 5) for ~ in with break

* break: 반복문 종료

```
for i in range(1, 10):
    if i > 5:
        break
    print(i)
```

> ## 6) for ~ in with continue

* continue: 반복문의 처음으로 돌아감

```
score = [99, 88, 55, 77, 66, 55, 100]

for num, s in enumerate(score):
    num = num + 1

    if s < 60:
        continue 
    print('%d번 수강생은 합격입니다.' % num)
```
