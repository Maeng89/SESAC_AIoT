# II. Python Data Type

> ## 1) Numeric

* 정수

```
print(888)
print(type(888))
```

* 8진수

```
print(0o77)
print(type(0o77))
```

* 16진수

```
print(0xff)
print(type(0xff))
```

* 2진수

```
print(0b1001)
print(type(0b1001))
```

* 실수

```
print(3.14)
print(type(3.14))
```

* 과학적 표기법

```
print(5e-4)
print(type(5e-4))
```

```
print(5e+4)
print(type(5e+4))
```

* 복소수(Complex)

```
print(8 + 9j)
print(type(8 + 9j))
```

> ## 2) String

```
print('A')
print(type('A'))
```

```
print('Data Analytics')
print(type('Data Analytics'))
```

> ## 3) Logical

* 참(True)
* 거짓(False)

```
print(True)
print(type(True))
```

```
print(False)
print(type(False))
```
