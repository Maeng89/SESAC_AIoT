```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from time import sleep
import random
import pandas as pd

options = webdriver.ChromeOptions()
# start = '202101011'
# end = '20211231'
# where = 'blog'
where = 'blog'
query = '인공지능'

user_agent = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome102.0.5005.115 Safari/537.36"
options.add_argument('user-agent=' + user_agent)
options.add_argument("window-size=1200,800")
driver = webdriver.Chrome('./chromedriver.exe', options = options)

#네이버 뷰는 최대 1050까지만 노출
```

```python
def get_view():
    lt = driver.find_element(By.CSS_SELECTOR, 'ul.lst_total')
    views = lt.find_elements(By.XPATH, 'child::li')
    result = []
    for view in views:
        title = view.find_element(By.CSS_SELECTOR, 'a.api_txt_lines.total_tit').get_attribute('innerText')
        text = view.find_element(By.CSS_SELECTOR, 'div.api_txt_lines.dsc_txt').get_attribute('innerText').replace('...','')
        date = view.find_element(By.CSS_SELECTOR, 'span.sub_time.sub_txt').get_attribute('innerText')[:-1]
        
        result.append({
            'title': title,
            'text': text,
            'date': date
        })
    return result
```

```python
dateform = '20221001to20221101'


fname = '{}_{}_{}'.format(where, query, dateform)

url = 'https://search.naver.com/search.naver?query={}&nso=so%3Ar%2Cp%3Afrom{}%2Ca%3Aall&where={}&sm=tab_opt'\
    .format(query,dateform,where)

driver.get(url)
sleep(random.randint(2,3))
```
