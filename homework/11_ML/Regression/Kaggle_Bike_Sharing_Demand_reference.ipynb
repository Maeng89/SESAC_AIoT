# Data Load
* 'Bike_Sharing_Demand.csv' from Github

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/rusita-ai/pyData/master/Bike_Sharing_Demand.csv'
DF = pd.read_csv(url)

DF.info()
```

```python
DF.head(3)
```

# I. Data Preprocessing

> ## 1) 'object' -> 'datetime64'

```python
DF['datetime'] = pd.to_datetime(DF['datetime'])
```

```python
DF.info()
```

> ## 2) 'year', 'month', 'day', 'hour', 'weekday' 추출

```python
DF['year'] = DF['datetime'].dt.year
DF['month'] = DF['datetime'].dt.month
DF['day'] = DF['datetime'].dt.day
DF['hour'] = DF['datetime'].dt.hour
DF['weekday'] = DF['datetime'].dt.weekday
```

```python
DF.info()
```

```python
DF.head(3)
```

> ## 3) Drop Columns

- 'datetime'

```python
DF.drop(columns = 'datetime', axis = 1, inplace = True)
```

```python
DF.head(3)
```

> ## 4) 분석용 DataFrame 백업

```python
DF0 = DF.copy()

DF0.head(3)
```

> ## 5) 시각화 레이블 변환

```python
DF['season'] = DF['season'].map({1: 'Spring', 
                                 2: 'Summer', 
                                 3: 'Fall', 
                                 4: 'Winter'})

DF['weather'] = DF['weather'].map({1: 'Clear', 
                                   2: 'Mist, Few clouds', 
                                   3: 'Light Snow, Rain, Thunder', 
                                   4: 'Heavy Snow, Rain, Thunder'})

DF['holiday'] = DF['holiday'].map({0: 'WorkingDay', 
                                   1: 'Holiday'})

DF['workingday'] = DF['workingday'].map({0: 'Holiday', 
                                         1: 'WorkingDay'})

DF['weekday'] = DF['weekday'].map({0: 'Monday', 
                                   1: 'Tuesday', 
                                   2: 'Wednesday', 
                                   3: 'Thursday',
                                   4: 'Friday',
                                   5: 'Saturday',
                                   6: 'Sunday'})
```

```python
DF.head(3)
```

# III. Modeling - Multivariate Regression

> ## 1) 변수 선택 - 'casual', 'registered' 제외
- count = casual + registered 이기 때문 -> Data Leakage

* X : 'month', 'day', 'hour', 'season', 'workingday', 'weather', 'atemp'
* y : 'count'

```python
DF1 = DF0[['month', 'day', 'hour', 'weekday', 'season', 'workingday', 'weather', 'atemp', 'count']].copy()

DF1.head(3)
```

> ## 2) train_test_split( )

* 7 : 3
 - random_state = 2045

```python
from sklearn.model_selection import train_test_split

X = DF1[['month', 'day', 'hour', 'weekday', 'season', 'workingday', 'weather', 'atemp']]
y = DF1['count']

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size = 0.3,
                                                    random_state = 2045)
print('Train Data : ', X_train.shape, y_train.shape)
print('Test Data : ', X_test.shape, y_test.shape)
```

> ## 3) 모델 생성

```python
import sklearn
from sklearn.linear_model import LinearRegression
sklearn.__version__
```

```python
#%%time # 동작 시간 매직코드, 리눅스에서만 사용 가능

from sklearn.linear_model import LinearRegression

# MR = LinearRegression(normalize = True, n_jobs = -1) # n_jobs cpu 몇개를 쓸 것인지, -1은 전체를 쓰겠다.
MR = LinearRegression(n_jobs = -1) # noremalize 매개변수는 사이킷런1.2부터 제거됨

MR.fit(X_train, y_train)
```

> ## 4) 모델 평가

```python
DF['count'].mean()
```

```python
from sklearn.metrics import mean_squared_error

np.sqrt(mean_squared_error(y_test, MR.predict(X_test)))
```

```python
from sklearn.metrics import r2_score

r2_score(y_test, MR.predict(X_test))
```

# V. Modeling - Log Transformation

> ## 1) 로그 변환 전

```python
# 편향된 데이터(y )에 대한 예측은 매우 어렵다.
# 스케일링은 주로 x값에 사용하고 log변환은 y값에 주로 사용
# 스케일링은 범위 변환이고 분포는 거의 바뀌지 않는다.
# x값의 분포를 바꾸는 것은 데이터의 특징이 왜곡 될 수 있는 위험이 있다
# 정규분포로 데이터가 수집되면 예측이 더 잘 될 수는 있겠지만 데이터의 특징을 잃어버린다.
```

```python
sns.displot(DF['count'], height = 6, aspect = 2)
plt.show()
```

> ## 2) 로그 변환 후

```python
sns.displot(np.log(DF['count']), height = 6, aspect = 2)
plt.show()
```

> ## 3) 'count' 로그 변환

```python
DF2 = DF0[['month', 'day', 'hour', 'weekday', 'season', 'workingday', 'weather', 'atemp', 'count']].copy()

DF2.head(3)
```

```python
DF2['count'] = np.log(DF2['count'])
```

```python
DF2.head(3)
```

> ## 4) train_test_split( )

* 7 : 3
 - random_state = 2045

```python
from sklearn.model_selection import train_test_split

X = DF2[['month', 'day', 'hour', 'weekday', 'season', 'workingday', 'weather', 'atemp']]
y = DF2['count']

X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size = 0.3,
                                                    random_state = 2045)
print('Train Data : ', X_train.shape, y_train.shape)
print('Test Data : ', X_test.shape, y_test.shape)
```

> ## 5) 모델 생성

```python
#%%time

from sklearn.linear_model import LinearRegression

# MR = LinearRegression(normalize = True, n_jobs = -1)
MR = LinearRegression(n_jobs = -1)

MR.fit(X_train, y_train)
```

> ## 6) 모델 평가

```python
from sklearn.metrics import mean_squared_error

np.sqrt(mean_squared_error(y_test, MR.predict(X_test)))
```

```python
from sklearn.metrics import r2_score

r2_score(y_test, MR.predict(X_test))
```
